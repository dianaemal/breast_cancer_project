```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("/Users/dianaemal/breast_cancer_project/results/model_results.csv")
data = pd.DataFrame(df)
print(data)
```

```python
data.loc[1, "Model"] = "Logreg(t = 0.3)"
models = data["Model"].astype(str)

accuracy = data["accuracy"]
recall = data["recall"]
precision = data["precision"]
f1 = data["F1"]

# Colors and labels
bar_colors = ['tab:purple', 'tab:blue', 'tab:red', 'tab:orange'][:len(models)]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()  # flatten to loop over easily

# Metric data and titles
metrics = [accuracy, recall, precision, f1]
titles = ["Accuracy", "Recall", "Precision", "F1 Score"]

for i in range(4):
    axs[i].bar(models, metrics[i], color=bar_colors)
    axs[i].set_title(f"{titles[i]} Comparison")
    axs[i].set_ylabel(titles[i])
    
plt.savefig("/Users/dianaemal/breast_cancer_project/results/comparisons.png", dpi=300)
plt.tight_layout()
plt.show()
```
