```python
import os
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
```

```python
# Load the model:
# Use model that is pretrained on imagenet datase
# include_top = False will freeze the prediction layer because we only want feature extraction
model = VGG16(weights='imagenet', include_top= False)
print(model.summary())
```

```python
path = "/Users/dianaemal/breast_cancer_project/data/raw"
b_images = os.listdir(f"{path}/benign")

x_features = []
y_labels = []
for image in b_images:
     if image.endswith(".png"):
        image_path = f"{path}/benign/{image}"

   

        # Load image and change the size because that's the size our model expects:
        img = load_img(image_path, target_size=(224, 224))
        # Convert pixels to numpy array:
        img = img_to_array(img)
        # Our model expects an input of 4 dimentions: (batch_size, height, width, RGB). currently we have three dimentions so we will
        #have to add the batch size and reshabe the image.
        img = np.expand_dims(img, axis=0)
        img =  preprocess_input(img)
        feature = model.predict(img)
        x_features.append(feature)
        y_labels.append(0)

    
    
```

```python
m_images =  os.listdir(f"{path}/malignant")

for image in m_images:
    if image.endswith(".png"):
        image_path = f"{path}/malignant/{image}"
        img = load_img(image_path, target_size=(224,224))
        img = img_to_array(img)
        img = np.expand_dims(img, axis=0)
        img = preprocess_input(img)
        feature = model.predict(img)
        x_features.append(feature)
        y_labels.append(1)
```

```python
print(len(y_labels))
```

```python
print(len(x_features))
```

```python
from sklearn.utils import shuffle
# Since we ran two loops seperetly on malignant and benign, the features and labels are save in order (first benign then malignant).
# We need to mix the values using shuffle:
x_features, y_labels = shuffle(x_features, y_labels, random_state=50)
# Change the lists into numpy arrays:
x_features = np.array(x_features)
y_labels = np.array(y_labels)
print(y_labels)
```

```python
# Save the arrays:
np.save('/Users/dianaemal/breast_cancer_project/data/features/X_features.npy', x_features)
np.save('/Users/dianaemal/breast_cancer_project/data/features/y_labels.npy', y_labels)
```
